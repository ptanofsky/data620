```python
unq_route_id = df_stop_times['route_id'].unique()
len(unq_route_id) # 28 unique route_id 
```

```python
def getLongestStopSeq(route_id):
    df_route_id = df_stop_times_temp[df_stop_times_temp['route_id'] == route_id]
    column = df_route_id['stop_sequence']
    max_value = column.max()
    results = df_route_id[df_route_id['stop_sequence']==max_value]['trip_id']
    results.reset_index(drop=True,inplace=True)
    trip_id = results[0]
    return trip_id
```

```python
# this gets the longest trip based on stop_sequence for a route_id, and find the trip_id, which is the name
res = getLongestStopSeq('2')
res
```

```python
longest_seq_per_route_id = []
for x in unq_route_id:
    trip_id = getLongestStopSeq(x)
    trip_seq_pair = (x, trip_id)
    longest_seq_per_route_id.append(trip_seq_pair)
    
longest_seq_per_route_id
```

```python
df_longest_seq_per_route_id = pd.DataFrame(longest_seq_per_route_id)
df_longest_seq_per_route_id.columns = ['route_id', 'trip_id']
df_longest_seq_per_route_id
```

```python
df_longest_trips = pd.DataFrame()
for idx, row in df_longest_seq_per_route_id.iterrows():
    df_longest_route_seq = df_stop_times_temp[df_stop_times_temp['trip_id'] == row['trip_id']]
    df_longest_route_seq = df_longest_route_seq.sort_values(by=['stop_sequence'])
    df_longest_route_seq['stop_id'] = df_longest_route_seq['stop_id'].str.replace('S', '')
    df_longest_route_seq['stop_id'] = df_longest_route_seq['stop_id'].str.replace('N', '')
    df_longest_trips = df_longest_trips.append(pd.DataFrame(data = df_longest_route_seq), ignore_index=True)
    
df_longest_trips
```

```python
# *** Above is where I'm at, one DF with the longest seq for each route_id, thus it should be 817-28 edges
# For each route, create the edges, and include the route_id
# loop through the each route longest seq and create a DF of the nodeA and nodeB
stop_pairs = []
for x in unq_route_id:
    df_longest_trips_per_route = df_longest_trips[df_longest_trips['route_id'] == x]
    df_longest_trips_per_route.reset_index(drop=True,inplace=True)
    for i, row in df_longest_trips_per_route.iterrows():
        stopA = ''
        stopB = ''
        if i > 0:
            stopA = df_longest_trips_per_route.loc[i-1, 'stop_id']
            stopB = df_longest_trips_per_route.loc[i, 'stop_id']
            stop_pair = (stopA, stopB, x)
            stop_pairs.append(stop_pair)
            
len(stop_pairs)
```

```python
# 789 edges but now I need to dedup those edges
# stopA, stopB, weight
```

```python
pd.set_option('display.max_rows', 800)
df_stop_pairs = pd.DataFrame(stop_pairs)
df_stop_pairs.columns = ['stop_a', 'stop_b', 'route_id']
df_stop_pairs
```

```python
# Ignore this block, as groupby and count was sufficient
column_names = ["stop_a", "stop_b", "weight"]
stop_pairs_dedup = pd.DataFrame(columns = column_names)

for i, row in df_stop_pairs.iterrows():
    if row['stop_a'] in stop_pairs_dedup['stop_a'] and row['stop_b'] in stop_pairs_dedup['stop_b']:
        # increment weight
        idx = stop_pairs_dedup.index(row)
        row_pop = stop_pairs_dedup.pop(idx)
        row_pop[-1] = row_pop[-1] + 1
        stop_pairs_dedup.append(row_pop)
        print('if')
    elif row['stop_a'] in stop_pairs_dedup['stop_b'] and row['stop_b'] in stop_pairs_dedup['stop_a']:
        # increment weight
        row_reverse = (row[1],row[0],row[2])
        idx = stop_pairs_dedup.index(row_reverse)
        row_pop = stop_pairs_dedup.pop(idx)
        row_pop[-1] = row_pop[-1] + 1
        stop_pairs_dedup.append(row_pop)
        print('elif')
    else:
        df = pd.DataFrame([[row['stop_a'], row['stop_b'], 1]], columns=['stop_a', 'stop_b', 'weight'])
        stop_pairs_dedup = pd.concat([stop_pairs_dedup, df])
(stop_pairs_dedup)
```

```python
# https://stackoverflow.com/questions/64605551/how-can-i-group-by-two-columns-interchangeably
df_stop_pairs_weight = df_stop_pairs[['stop_a', 'stop_b']]
df_stop_pairs_weight['tmp'] = df_stop_pairs_weight.apply(frozenset, axis=1)
df_stop_pairs_weight['weight'] = df_stop_pairs_weight.groupby('tmp')['stop_a'].transform('size')
df_stop_pairs_weight = df_stop_pairs_weight.set_index('tmp')
df_stop_pairs_weight = df_stop_pairs_weight[~df_stop_pairs_weight.index.duplicated()].reset_index(drop=True)
df_stop_pairs_weight
```
