## Retrieve voteview.com data and display at dataframe

```python
# Import libraries as needed
import urllib.request, json 
import pandas as pd
```

```python
# Retrieve data from URL in JSON format
with urllib.request.urlopen("https://voteview.com/api/download?rollcall_id=RS1170600") as url:
    data = json.loads(url.read().decode())
```

```python
# Extract vote data
votes = data['rollcalls'][0]['votes']
# Extract vote title string
vote_title = data['rollcalls'][0]['vote_title']
# Convert vote data to Pandas DF
df = pd.DataFrame(votes)
# Output DF
df.head(5)
```

```python
# Trim DF by column name
df1 = df[['seo_name','party_short_name', 'state', 'vote']]
# Rename column header, as I would expect we'd have each bill to be a column
# I suppose, we would actually want to convert 'Yea' to 1, and everything else to '0'
df1 = df1.rename(columns={'vote': 'vote_title'})
df1
```
